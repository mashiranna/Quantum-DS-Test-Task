3. You have a dataset (internship_train.csv) that contains 53 anonymized features and a target column. Your task is to build model that predicts a target based on the proposed features. Please provide predictions for internship_hidden_test.csv file. Target metric is RMSE. The main goal is to provide github repository that contains:
jupyter notebook with analysis;
code for modeling (Python 3);
file with model predictions;
readme file;
requirements.txt file.

```python
import pandas as pd
import numpy as np
import math
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
```

```python
train_data = pd.read_csv('internship_train.csv')
test_data = pd.read_csv('internship_hidden_test.csv')
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
number_of_rows, number_of_cols = train_data.shape
print("Number of rows: ", number_of_rows)
print("Number of columns: ", number_of_cols)
print("\nData type of each column: ")
print(train_data.dtypes)
```

```python
target_data = train_data[['target']].copy()
train_data = train_data.drop(['target'], axis=1)
```

```python
x_train, x_true, y_train, y_true = train_test_split(train_data, target_data, test_size=0.25)
```

 Linear Regression

```python
from sklearn.linear_model import LinearRegression

clf = LinearRegression(fit_intercept = True, copy_X = True, positive = True).fit(x_train, y_train)
predicted = clf.predict(x_true)

print(clf.score(x_true, y_true))

# The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse)
```

```python
MSE = mean_squared_error(y_true, predicted)
RMSE = math.sqrt(MSE)
print(f'Root Mean Square Error: {RMSE}')
```

Decision Tree Regression
Decision tree regression is a non-parametric model that fits a tree structure to the data. It can capture complex nonlinear relationships between features and target variables, but may be prone to overfitting.

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(max_depth=7, random_state=1).fit(x_train, y_train)
predicted = regressor.predict(x_true)

print(regressor.get_depth())
print(regressor.get_n_leaves())
print(regressor.score(x_true, y_true))
```

```python
print('Feature importances\n', list(zip(regressor.feature_importances_, np.arange(53))))
```

RMSE

```python
MSE = mean_squared_error(y_true, predicted)
RMSE = math.sqrt(MSE)
print(f'Root Mean Square Error: {RMSE}')

# RMSD is always non-negative, and a value of 0 
# (almost never achieved in practice) would indicate a 
# perfect fit to the data. In general,
# a lower RMSD is better than a higher one
```

```python
prediction_for_test_data = regressor.predict(test_data)
```

```python
prediction = pd.DataFrame(prediction_for_test_data, columns=['predictions']).to_csv('prediction.csv')
```
